```python
import pandas as pd
import ast  # برای تبدیل رشته به ساختار پایتونی

path = "/media/mohammad/NewVolume/code/result_x/loso.csv"

df = pd.read_csv(path, sep=";", on_bad_lines="skip")

# تابع تبدیل هر سلول Param به دیکشنری
def parse_param_cell(cell):
    if not isinstance(cell, str):
        return {}
    s = cell.strip()
    # اگر با " شروع و تمام می‌شود (مثل خروجی CSV)
    if s.startswith('"') and s.endswith('"'):
        s = s[1:-1]
    try:
        data = ast.literal_eval(s)   # تبدیل رشته به list/tuple/dict پایتونی
    except Exception as e:
        print("BAD CELL:", cell, e)
        return {}
    
    # اگر لیستِ tuple است، تبدیلش کن به dict
    if isinstance(data, list):
        try:
            data = dict(data)
        except Exception:
            pass
    return data if isinstance(data, dict) else {}

# ۱) تبدیل ستون Param به دیکشنری برای هر ردیف
param_dicts = df["Param"].apply(parse_param_cell)

# ۲) باز کردن دیکشنری‌ها به ستون‌های جدا
param_df = param_dicts.apply(pd.Series)

# ۳) چسباندن ستون‌های جدید به df اصلی
df_expanded = pd.concat([df, param_df], axis=1)

# اگر دیگه ستون Param قدیمی رو نمی‌خوای:
# df_expanded = df_expanded.drop(columns=["Param"])

print(df_expanded.head())
print(df_expanded.columns)
```

```python
def show_data(slit_calsses):
    # ۱) شرط: هر سطری که classes دقیقا [[0], [1]] باشد
    mask = df_expanded["classes"].apply(lambda x: x == slit_calsses)

    # ۲) فیلتر کردن و فقط ستون‌های مورد نظر
    result = df_expanded.loc[mask, ["Accuracy mean", 'Accuracy std' , 'F1 mean','n_estimators']]

    print(result)
```
